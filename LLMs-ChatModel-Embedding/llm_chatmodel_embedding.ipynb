## Chat-Model With  GPT implementation
current method

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

model = ChatOpenAI(model = "gpt-5.6-luna", temperature = 0.7, max_completion_tokens=50)

result = model.invoke("Explain machine learning in simple terms.")

print(result.content)   
```

## Chat-Model With  Anthropic implementation

```python
from langchain_anthropic import ChatAnthropic
from dotenv import load_dotenv

load_dotenv()

model = ChatAnthropic(
    model="claude-sonnet-4",
    temperature=0.7,
    max_tokens=50
)

result = model.invoke("Explain machine learning in simple terms.")

print(result.content)
```

## Open Source Hugging Face implementation through API 
it is also paid but give some free tokens

### Access:
click on profile >> click on Access Token >> create API token >> Read/Write 

AT .env => HF_TOKEN=.........put-api........

### for llm
go to : model >>  Task >> Text-genetation >> choose any: meta-llama/Llama-3.1-8B-Instruct

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv

load_dotenv()

llm = HuggingFaceEndpoint(
    repo_id = 'meta-llama/Llama-3.1-8B-Instruct',
    task = 'text-generation'
)

model = ChatHuggingFace(llm=llm)

result = model.invoke("what is capital of India")
print(result.content)
```

## Open Source Hugging Face implementation on local machine\laptop

```python
from langchain_huggingface import  ChatHuggingFace, HuggingFacePipeline

llm = HuggingFacePipeline.from_model_id(
    model_id= 'TinyLlama/TinyLlama-1.1B-Chat-v1.0',
    task='text-generation',
    pipeline_kwargs = dict(
        temperature=0.5,
        max_new_tokens=100
    )
)
model = ChatHuggingFace(llm=llm)

result = model.invoke("what is capital of India")
print(result.content)
```
